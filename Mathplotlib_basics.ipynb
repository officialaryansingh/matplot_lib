# Matplotlib:

  .Useful for making Plots

```python
# importing matplotlib library

import matplotlib.pyplot as plt

# importing numpy to get our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Ploting the data

```python
# sin wave

plt.plot(x,y)
plt.xlabel('x')#this give the labeling of the x axis in the plot
plt.ylabel('sin wave')
plt.title('sin wave function')
plt.show()
```
