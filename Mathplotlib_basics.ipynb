# Matplotlib:

  .Useful for making Plots

```python
# importing matplotlib library

import matplotlib.pyplot as plt

# importing numpy to get our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Ploting the data

```python
# sin wave

plt.plot(x,y)
plt.xlabel('x')#this give the labeling of the x axis in the plot
plt.ylabel('sin wave')
plt.title('sin wave function')
plt.show()
```

```python
# parabola

x = np.linspace(-10,10,25)
y = x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('parabola y = x^2')
plt.title('parabola function')
plt.show()
```

```python

# changing the layout out of the output
plt.plot(x,y,'g.') # g - green for colour and . - for layout
```

```python
# ploting multiple function in one plot

x= np.linspace(-10,10,100)
plt.plot(x,np.sin(x),'g-')

plt.plot(x,np.cos(x),'r--')

# y = x**2
# plt.plot(x, y, 'b*')

plt.plot()
```

#Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','German','Latin']
people = [100,50,150,30,20]
ax.bar(languages,people)
plt.xlabel('LANGUAGE')
plt.ylabel('PEOPLE')
plt.show()
```

#Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','German','Latin']
people = [200,50,40,150,20]
ax.pie(people, labels = languages, autopct = '%1.2f%%')
plt.show()
```

#Scatter PLot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color = 'g')

ax.scatter(x,z,color = 'b')

plt.show()
```

#3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')

z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap= 'Blues')
plt.show()
```

```python
fig4 = plt.figure()
ax = plt.axes(projection = '3d')
z = np.linspace(0,100,500)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c = z, cmap = "Reds")
plt.show()
```
